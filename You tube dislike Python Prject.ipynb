```python
yt['published_month']=ytd['published_at'].apply(lambda x:x[5:7])
yt['published_month'].unique()
```

# Q7

```python
yt['published_month']=yt['published_month'].replace('01','Jan')
yt['published_month']=yt['published_month'].replace('02','Feb')
yt['published_month']=yt['published_month'].replace('03','March')
yt['published_month']=yt['published_month'].replace('04','April')
yt['published_month']=yt['published_month'].replace('05','May')
yt['published_month']=yt['published_month'].replace('06','June')
yt['published_month']=yt['published_month'].replace('07','July')
yt['published_month']=yt['published_month'].replace('08','Aug')
yt['published_month']=yt['published_month'].replace('09','Sep')
yt['published_month']=yt['published_month'].replace('10','Oct')
yt['published_month']=yt['published_month'].replace('11','Nov')
yt['published_month']=yt['published_month'].replace('12','Dec')
yt['published_month'].unique()
```

```python
month=pd.pivot_table(data=yt[['published_month','video_id']],index='published_month',aggfunc='count')
month.sort_values(by='video_id',ascending=False)
```
